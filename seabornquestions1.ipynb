Q-2-3: Using flights dataset of seaborn.
1.Using the "flights" dataset that comes with seaborn, create a heatmap that shows the average number of passengers per month for each year.

2.Using the seaborn's flight dataset, create a clustermap to visualize the relationship between the number of passengers, months, and year.

```python
import seaborn as sns 
tips=sns.load_dataset('tips')
iris=sns.load_dataset('iris')
```

```python
tips.head()
```

```python
#strip plot is a axes level function which can be used to plot between a categorical column and a numerical column
import seaborn as sns 
sns.stripplot(data=tips,x='day',y='total_bill')
```

```python
import seaborn as sns 
sns.stripplot(data=tips,x='day',y='total_bill',jitter=False)
#we do not get a straight line as jitter by default is True and there is noise, when jitter is made False then we get a straight
#line. 
```

```python
#catplot is a figure level function 
import seaborn as sns 
sns.catplot(data=tips,x='day',y='total_bill',kind='strip') 
```

```python
import seaborn as sns 
sns.catplot(data=tips,x='day',y='total_bill',jitter=0.1,kind='strip') 
#catplot is a figure level function and the jitter parameter within it can be used to adjust the amount of jitterness
```

```python
#jitter 
import seaborn as sns 
sns.catplot(data=tips,x='day',y='total_bill',kind='strip',jitter=0.2,hue='sex')
```

```python
import seaborn as sns 
sns.catplot(data=tips,x='day',y='total_bill',kind='swarm')
```

```python
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
sns.swarmplot(data=tips,x='day',y='total_bill')
```

Boxplot


A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

```python
#boxplot
import seaborn as sns 
sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
#boxplot using catplot 
import seaborn as sns 
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='box')
```

```python
#single boxplot can be done on a numerical column 
import seaborn as sns 
sns.boxplot(data=tips,y='total_bill')
```
